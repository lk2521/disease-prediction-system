```python
import shutil
shutil.move('Testing.csv', os.path.join('dataset', 'Testing.csv'))
shutil.move('Training.csv', os.path.join('dataset', 'Training.csv'))
```

```python
import pandas as pd
```

```python
dataset = './dataset'
```

```python
train_df = pd.read_csv(dataset + '/Training.csv')
```

```python
test_df = pd.read_csv(dataset + '/Testing.csv')
```

```python
train_df.drop(train_df.columns[-1], axis=1, inplace=True)
```

```python
X = train_df.drop(labels = 'prognosis',axis=1)
y = train_df['prognosis']
```

```python
X_test = test_df.drop(labels = 'prognosis',axis=1)
y_test = test_df['prognosis']
```

```python
X
```

```python
y
```

```python
from sklearn.svm import SVC
```

```python
model = SVC()
```

```python
model.fit(X,y)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score,classification_report
```

```python
accuracy_score(y_test, y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```
